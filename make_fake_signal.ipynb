```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-25, 25, 200)

def gaussian(x, mean, std):
    return np.exp(-(x-mean)**2 / (2*std**2))

gaussian1 = gaussian(x, -4, 0.8)
gaussian2 = gaussian(x, 4, 0.8)

head = gaussian(x, 0, 4)

signal = gaussian1 + gaussian2 + head

plt.figure(figsize=(8, 4))
plt.plot(x, signal)
plt.fill_between(x, signal, color='gray', alpha=0.5)
plt.title('Scary Ghost Signal')
plt.legend()
plt.show()


np.save('true_signal.npy', signal)
```

```python
from scipy.special import voigt_profile 

# this is the profile that will blur our observation
kernel = voigt_profile(np.linspace(-25, 25, 200), 2., 0.5)
kernel /= kernel.sum()

obs = np.convolve(signal, kernel, mode='same')
plt.figure(figsize=(8, 4))
plt.plot(x, obs)
plt.fill_between(x, obs, color='gray', alpha=0.5)
plt.title('Not so scary now')
plt.legend()
plt.show()


np.save('kernel.npy', kernel)
```
